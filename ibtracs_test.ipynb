```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
fl='ibtracs.NA.list.v04r00_USA.csv';
df=pd.read_csv(fl,skiprows = [1]).drop(['SID','DIST2LAND','LANDFALL','USA_LAT','USA_LON','USA_STATUS','USA_AGENCY','IFLAG','TRACK_TYPE','WMO_AGENCY','WMO_WIND','WMO_PRES','USA_SEAHGT','USA_SEARAD_SW','USA_SEARAD_NW','USA_SEARAD_NE','USA_SEARAD_SE'],axis=1).replace(' ', np.nan)
df.USA_PRES=pd.to_numeric(df.USA_PRES)
df.USA_WIND=pd.to_numeric(df.USA_WIND)
df.USA_RMW=pd.to_numeric(df.USA_RMW)
```

```python
df.keys()
```

```python
fig = plt.figure(figsize=(16, 10))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()
ax.coastlines()

df.plot.scatter(x='LON',
                y='LAT',
                c='USA_RMW',
                ax=ax)
```
